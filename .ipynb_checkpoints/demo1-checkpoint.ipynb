## 1.1 Comparison of Different Affinity Matrix

The affinity matrix are vital to the success of a graph-cut algorithm. Some commonly used affinity rules are: `0-1 nearest neighbors` , `rbf`, `cosine`. Given an affinity matrix $W$, we can build a symmetric graph Laplacian $L_s = I- D^{1/2}WD^{1/2}$ and compute its smallest eigenvectors $V$. 

A spectral embedding is the projection of $x$ onto $\mathbb{R}^k$ using the eigenvectors of the graph Laplacian $V$. This embedding can often transform non-linear, non-convex clusters into convex and linearly separable clusters. The spectral clustering algorithm simply applies `K-means` to the transformed domain

We will plot the spectral embedding of the three "twomoons" dataset using different graph kernels. Below is a plot for nearest neighbors

```python
# 0-1 nearest neighbor graph
plt.figure(2)
plt.rcParams['figure.figsize'] = (12, 14)
for i in range(3):
    for j in range(4):
        lap = util.build_laplacian_matrix(raw_data = x[:,:,i],affinity = 'nearest_neighbors', n_neighbors = 15+13*(j))
        E,V = eigsh(lap,3,which = 'SM')
        x_t = V[:,1]
        y_t = V[:,2]
        color = 'r'*n_mid + 'b'*n_mid
        plt.subplot(3,4,i*4+j+1)
        plt.scatter(x_t,y_t,s = 3, color = color)
        plt.axis('off')
        plt.title('NN, K = {K}, Sig = {sigma}'.format(K = 15+13*(j), sigma = 0.02+.03*(i+1)))
plt.show()
```

We can see from the embeddings that spectral clustering will work well with the nearest neighbor graphs. We will try out the  `rbf` kernel next which is also frequently used. 

```python
# RBF kernel graph
plt.figure(3)
plt.rcParams['figure.figsize'] = (12, 14)
for i in range(3):
    for j in range(4):
        kernel_params = {'gamma': .4+.8*j}
        lap = util.build_laplacian_matrix(raw_data = x[:,:,i],affinity = 'rbf',kernel_params = kernel_params)
        E,V = eigsh(lap,3,which = 'SM')
        x_t = V[:,1]
        y_t = V[:,2]
        color = 'r'*n_mid + 'b'*n_mid
        plt.subplot(3,4,i*4+j+1)
        plt.scatter(x_t,y_t,s = 3, color = color)
        plt.axis('off')
        plt.title('gamma = {g}, Sig = {sigma}'.format(g = .4+.5*j, sigma = 0.02+.03*(i+1)))
plt.show()
```

We see that the non-linearities are not at all resolved for $\gamma = .4$, and significant overlap exists for larger $\gamma$s. It's fair to conclude that the `rbf` kernel is not suited for segmenting this problem

## 1.2 Demo of Semi-Supervised Graph Cut

We will demonstrate how to use the graph clustering package and show some results

First off, we will import the module and initialize the the classifier

```python
#initialization
import graph_cluster as gc
reload (gc)
params = {'eps':1, 'eta':2} # parameters for the classifier
gl_clf = gc.LaplacianClustering(scheme_type = 'GL_fidelity',  params = params)
```

The scheme chosen was a Ginzburg-Landau scheme with fidelity term. Parameters are set in the dictionary params. 

Next, we load the data into the classifier via `load_raw_data` function. Ground truth is optional, but can be used to calculate miss-classification and also generate random fidelity. 

```python
# load raw_data
x_train = x[:,:,1]
mid = x_train.shape[0]/2
ground_truth = np.concatenate((np.ones(mid),-np.ones(mid)),axis = 0)
gl_clf.load_raw_data(raw_data = x_train, ground_truth = ground_truth)
gl_clf.generate_random_fidelity(percent = .1)
gl_clf.generate_initial_value() #random initial condition generated
```

We then have to specify the graph specs for the classifier via `set_graph_parameters` function. The options are the same as in the `sklearn` package. 

```python
gl_clf.set_graph_parameters(affinity = 'nearest_neighbors', n_neighbors = 30, Neig = 50)
```

Finally, we simply fit the data at hand and plot the predictions

```python
# Testing supervised learning
plt.figure(4)
plt.rcParams['figure.figsize'] = (15, 5)
def gen_colors(u):
    color = ""
    for l in u:
        if l>0:
            color += 'r'
        else:
            color += 'b'
    return color
#classification results
for i in range(3):
    gl_clf.generate_random_fidelity(percent = (i+1)*.03)
    gl_clf.fit_predict()
    plt.subplot(1,3,i+1)
    plt.scatter(x_train[:,0],x_train[:,1],s = 3, color = gen_colors(gl_clf.labels_))
    plt.title('Fidelity Percent = {perc}'.format(perc = (i+1)*.03))    
    plt.axis('off')
plt.show()
```

Surprisingy, the number of eigenvectors is not monotone with the classification quality. It seems to decrease first and then increase as Neig gets larger. The user is welcome to try experimenting with different values of Neig. 

We will test the scheme `MBO fidelity` on the same problem. We use `set_parameters` to change the scheme type to 'MBO_fidelity'. Note neither the data nor the graph spec has to be changed.

```python
gl_clf.set_parameters(scheme_type='MBO_fidelity')
```

We perform the same experiment with the new scheme. 

```python
# Testing supervised learning
reload(gc)
gl_clf.set_graph_parameters(affinity = 'nearest_neighbors', n_neighbors = 40, Neig = 200)
plt.figure(5)
plt.rcParams['figure.figsize'] = (15, 5)
def gen_colors(u):
    color = ""
    for l in u:
        if l>0:
            color += 'r'
        else:
            color += 'b'
    return color
#classification results
for i in range(3):
    gl_clf.generate_random_fidelity(percent = (i+1)*.03)
    gl_clf.fit_predict()
    plt.subplot(1,3,i+1)
    plt.scatter(x_train[:,0],x_train[:,1],s = 3, color = gen_colors(gl_clf.labels_))
    plt.title('Fidelity Percent = {perc}'.format(perc = (i+1)*.03))    
    plt.axis('off')
plt.show()
```

Seems to do just as well. One remark is that the supervised `MBO` has an `inner loop count` constant that is not 1. Too low of a loop count does degrade the prediction error.

Same as before, we simply generate some fidelity and apply `fit_predict`. The graph is a nearest neighbor graph with $30$ neighbors
